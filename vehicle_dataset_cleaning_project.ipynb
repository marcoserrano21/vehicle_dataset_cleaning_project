# Data Cleaning project

A project to perform data cleaning tasks.

The "Vehicle Dataset 2024" dataset was found through Kaggle.com and is available under the ODC Attribution license.
https://www.kaggle.com/datasets/kanchana1990/vehicle-dataset-2024/data



**Column Descriptors**
* `name`: The full name of the vehicle, including make, model, and trim.
* `description`: A brief description of the vehicle, often including key features and selling points.
* `make`: The manufacturer of the vehicle (e.g., Ford, Toyota, BMW).
* `model`: The model name of the vehicle.
* `type`: The type of the vehicle, which is "New" for all entries in this dataset.
* `year`: The year the vehicle was manufactured.
* `price`: The price of the vehicle in USD.
* `engine`: Details about the engine, including type and specifications.
* `cylinders`: The number of cylinders in the vehicle's engine.
* `fuel`: The type of fuel used by the vehicle (e.g., Gasoline, Diesel, Electric).
* `mileage`: The mileage of the vehicle, typically in miles.
* `transmission`: The type of transmission (e.g., Automatic, Manual).
* `trim`: The trim level of the vehicle, indicating different feature sets or packages.
* `body`: The body style of the vehicle (e.g., SUV, Sedan, Pickup Truck).
* `doors`: The number of doors on the vehicle.
* `exterior_color`: The exterior color of the vehicle.
* `interior_color`: The interior color of the vehicle.
* `drivetrain`: The drivetrain of the vehicle (e.g., All-wheel Drive, Front-wheel Drive).

**Data Cleaning for duplicates**

```python
# Detection of Duplicates
duplicates = vehicles_df.duplicated()
dupl_counts = duplicates.value_counts()

print(f"Total number of duplicate rows: {duplicates.sum()} \n")
print(f"True vs. False values: \n{dupl_counts}")
```

```python
# Inspection of Duplicates for their descriptions
vehicles_df[vehicles_df.duplicated(keep=False)].sort_values('description').head()
```

```python
# Inspection of Duplicates for their name
vehicles_df[vehicles_df.duplicated(keep=False)].sort_values('name').head()
```

**Treatment of Duplicate Rows**

The `description` variable will be extracted into a separate dataset for later analysis since this variable can be useful for Natural Language processing. Removing the `description` variable may create additional duplicate rows. All duplicate rows will be dropped on the continued dataset.

```python
vehicle_only_df = vehicles_df.drop(columns='description', axis=1)

# Detection of duplicates cleaned dataframe
duplicates = vehicle_only_df.duplicated()
dupl_counts = duplicates.value_counts()

print(f"Total number of duplicate rows Before Dropping: {duplicates.sum()} \n")
print(f"True vs. False values: \n{dupl_counts}")

vehicle_only_df.drop_duplicates(inplace=True)

duplicates = vehicle_only_df.duplicated()
dupl_counts = duplicates.value_counts()

print(f"Total number of duplicate rows After Dropping: {duplicates.sum()} \n")
print(f"True vs. False values: \n{dupl_counts}")
```

**Detection of Missing Values**

```python
print(f'Total number of missing values: {vehicle_only_df.isna().sum().sum()}')
print(vehicle_only_df.isna().sum()) #Let's you see the missing values per column

# We'll add visualization of the missing values
import missingno as msno # Missingno provides a small toolset of flexible and easy-to-use missing data visualizations #

print(f'Data Sparsity: {vehicle_only_df.isna().sum().sum() / (vehicle_only_df.shape[0] * vehicle_only_df.shape[1])}')

msno.matrix(vehicle_only_df, labels=True, figsize=(14, 5))
plt.title('Visual Summary of the completeness of the dataset')
plt.show()
```

**Treatment of the Missing Values**



```python
# We'll drop the rows with missing values
vehicle_only_df.dropna(inplace=True)

# Check for missing values again to confirm the change
print(f'Total number of missing values: {vehicle_only_df.isna().sum().sum()}')
print(vehicle_only_df.isna().sum())


msno.matrix(vehicle_only_df, labels=True, figsize=(14, 5))
plt.title('Visual Summary of the completeness of the dataset')
plt.show()
```

# Visualizations

```python
# Distribution of Vehicle Prices
plt.figure(figsize=(10, 6))
sns.histplot(vehicle_only_df['price'], bins=50, kde=True, color='#94d2bd')
plt.title('Distribution of Vehicle Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

The histogram indicates that most vehicles are priced between $20,000 and $60,000, suggesting a concentration of vehicle prices within this range. The frequency decreases for prices above and below this range.
_______________

```python
# Price distribution by Vehicle Body Style
plt.figure(figsize=(10, 6))
sns.boxplot(x='body', y='price', data=vehicle_only_df, color='#ee9b00')
plt.title('Price Distribution by Vehicle Body Style')
plt.xlabel('Body Style')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.show()
```

The box plot shows vehicle price distributions by body style. SUVs and Pickup Trucks have the highest median prices, while Hatchbacks and Sedans are lower. There are significant price variations and outliers across all categories.
______________________

```python
# Distribution of Body Types
ax = sns.countplot(y='body', data=vehicle_only_df, color='#005f73')
for label in ax.containers:
    ax.bar_label(label)
plt.title('Distribution of Body Types')
plt.xlabel('Count')
plt.ylabel('Body Type')
plt.show()
```

Based on this chart, it shows the distribution of vehicle body types. SUVs are the most common with 532 units, followed by Pickup Trucks at 146. Other types like Sedans, Cargo Vans, and Hatchbacks have significantly lower counts. This highlights the popularity of SUVs in the market.
_________________

```python
# Mileage vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mileage', y='price', data=vehicle_only_df, alpha=0.5, color='#ef233c')
plt.title('Mileage vs. Price')
plt.xlabel('Mileage')
plt.ylabel('Price ($)')
plt.show()
```

The scatter shows that cars with lower mileage tend to have higher prices, with most data points clustered at the lower end of the mileage axis and higher on the price axis.
______________

```python
# Distribution of Fuel Types
sns.boxplot(x='price', y='fuel', data=vehicle_only_df, color='#9b2226')
plt.title('Distribution of Fuel Types')
plt.xlabel('Price ($)')
plt.ylabel('Fuel Type')
plt.show()
```

The box plot “Distribution of Fuel Types” shows that prices vary significantly across fuel types. Gasoline and Diesel are generally cheaper, while Electric and Hybrid vehicles tend to be more expensive. This highlights the price differences based on fuel type.
